```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# CA

## FTS

```python
df1 = pd.read_csv("cache/lr_data_fts_500.csv")
df2 = pd.read_csv("cache/simple_data_fts_500.csv")
df = pd.concat([df1, df2])
g = sns.lmplot(
    data=df,
    x="latent_dim", y="loss", hue="id",
    height=5,
    ci=50,
    # robust=True,
    legend=False,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
g.set_axis_labels(x_var="Latent Dimension", y_var="Loss")
plt.yscale('log')
plt.xscale('log')
plt.legend(title="Model")
plt.show()
```

## FTS

```python
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
df1 = pd.read_csv("cache/LDS/lr_data_fc_6_True.csv")
df1["id"] = "LR (in)"
df2 = pd.read_csv("cache/LDS/lr_data_fc_6_False.csv")
df2["id"] = "LR (out)"
df3 = pd.read_csv("cache/LDS/simple_data_6_True.csv")
df3["id"] = "DNN (in)"
df4 = pd.read_csv("cache/LDS/simple_data_6_False.csv")
df4["id"] = "DNN (out)"
df = pd.concat([
    df1,
    df2,
    df3,
    df4
])
_ = sns.regplot(
    ax=axs[0],
    data=df1,
    x="n", y="loss",
    label="LR (in)",
    ci=50,
    # robust=True,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
_ = sns.regplot(
    ax=axs[0],
    data=df2,
    x="n", y="loss",
    label="LR (out)",
    ci=50,
    # robust=True,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
_ = sns.regplot(
    ax=axs[0],
    data=df3,
    x="n", y="loss",
    label="DNN (in)",
    ci=50,
    # robust=True,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
_ = sns.regplot(
    ax=axs[0],
    data=df4,
    x="n", y="loss",
    label="DNN (out)",
    ci=50,
    # robust=True,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
axs[0].grid(visible=False)
axs[0].set_yscale('log')
axs[0].set_xscale('log')
axs[0].set_xlabel(r"$n$ samples")
axs[0].set_ylabel("MSE / Embed. Dim.")
axs[0].legend(title="Model")
axs[0].set_title("Fixed Complexity")
plt.tight_layout()
plt.show()
```
