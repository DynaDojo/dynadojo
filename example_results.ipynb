```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# CA

## FTS

```python
df1 = pd.read_csv("cache/lr_data_fts_500.csv")
df2 = pd.read_csv("cache/simple_data_fts_500.csv")
df = pd.concat([df1, df2])
g = sns.lmplot(
    data=df,
    x="latent_dim", y="loss", hue="id",
    height=5,
    ci=50,
    # robust=True,
    legend=False,
    # facet_kws=dict(sharey=False),
    scatter=False,
    fit_reg=True
)
g.set_axis_labels(x_var="Latent Dimension", y_var="Loss")
plt.yscale('log')
plt.xscale('log')
plt.legend(title="Model")
plt.show()
```
