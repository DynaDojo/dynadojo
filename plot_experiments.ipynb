```python
from experiments import make_plots
from dynadojo.challenges import FixedDimensionality, FixedTrainSize, FixedError
from dynadojo.utils.plotting import plot_metric
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
g1, data1 = make_plots(
    s ="lorenz",
    m = "lr",
    output_dir="experiments/outputs/scratch",
    challenge_cls = FixedTrainSize,
    save=False)
```

```python
g2, data2 = make_plots(
    s ="lorenz",
    m = "sindy",
    output_dir="experiments/outputs/scratch",
    challenge_cls = FixedTrainSize,
    save=False)
```

```python
y = pd.concat([data1, data2], axis=0, keys=['lr', 'sindy']).reset_index()
y = y.rename(columns={'level_0': 'algorithm', 'level_1': 'x'})
ax = sns.lineplot(data=y, x='latent_dim', y='error', hue='algorithm', linewidth=2.5, estimator=np.median, errorbar=("pi", 25))
ax.set(xscale="log")
ax.set(yscale="log")
ax.set(title="Fixed Train Size, n=1000")
ax.figure.savefig("fts_lorenz_lr_sindy.pdf", bbox_inches='tight')
```
